# Convolve Function

```python
def Convolve(img,kernel):
    img = np.asarray(img)
    h,w = img.shape[0],img.shape[1]
    kernel = np.flip(kernel)
    #################################################################################
    stride = 1
    new_i = 0
    fil_h = kernel.shape[0]
    fil_w = kernel.shape[1]
    new_img = np.zeros((h-fil_h+1,w-fil_w+1))
    #################################################################################
    
    for i in range(h - fil_h+1):
        for j in range(w - fil_w+1):
            im_region = img[i:(i + fil_h), j:(j + fil_w)]
            r = np.multiply(im_region,kernel).sum()
            new_img[i][j] = int(r/fil_h*fil_w)
    return new_img
```

# Convolve over 3d Image

```python
def Convolve3d(img,kernel):
    stacked_image = []
    stacked_image = np.asarray(stacked_image)
    stacked_image = Convolve(img[:,:,0],kernel)
    if len(img.shape) == 3:
        for i in range(1,3):
            #print(img1[:,:,i].shape)
            stacked_image = np.dstack((stacked_image,Convolve(img[:,:,i],kernel)))
            #stacked_image.append(Convolve(img1[:,:,i],kernel)Convolve(img1[:,:,i],kernel))
    return stacked_image
```

# Normalize array between 0-1

```python
def Normalizedbetweem01(kernel):
    return np.interp(kernel, (kernel.min(), kernel.max()), (0, 1))
```

# 2d gaussian function to generate the filter

```python
def gauss2D(shape=(15,15),sigma=1):
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h
```

# Number of times we have to iterate

```python
no_of_iter = 5
```

```python
mu, sigma = 0, 15 # mean and standard deviation

#kernel = np.random.(mu, sigma, (5,5))
kernel = gauss2D(shape=(5,5),sigma=sigma)
kernel = Normalizedbetweem01(kernel)#np.interp(kernel, (kernel.min(), kernel.max()), (0, 1))
```

```python
def paddingCalculate(O,F,S,W): #O : OutputSize , F : Filter , S : Stride , W : InputSize
    return ((O-1)*S + F - W)/2
```

# Low pass filter

```python
def lowPassFilter(img,kernel):
    stacked_image_dog = Convolve3d(img,kernel)
    for i in range(no_of_iter):
        #print(i)
        stacked_image_dog = Convolve3d(stacked_image_dog,kernel)
    stacked_image_dog = Normalizedbetweem01(stacked_image_dog)#np.interp(stacked_image_dog, (stacked_image_dog.min(), stacked_image_dog.max()), (0, 1))
    #stacked_image *= (255.0/stacked_image.max())
    plt.imshow(stacked_image_dog)
    plt.title("Low pass resulting Dog Image")
    plt.show()
    return stacked_image_dog
```

```python
stacked_image_dog = lowPassFilter(img1,kernel)
```

```python
def overlap(b1,b2,pos_v,pos_h):
    print(b1.shape)
    print(b2.shape)
#     b1 = np.zeros((4,5))
#     b2 = np.ones((4,4))
    pos_v, pos_h = pos_v, pos_h  # offset
    v_range1 = slice(max(0, pos_v), max(min(pos_v + b2.shape[0], b1.shape[0]), 0))
    h_range1 = slice(max(0, pos_h), max(min(pos_h + b2.shape[1], b1.shape[1]), 0))

    v_range2 = slice(max(0, -pos_v), min(-pos_v + b1.shape[0], b2.shape[0]))
    h_range2 = slice(max(0, -pos_h), min(-pos_h + b1.shape[1], b2.shape[1]))

    b1[v_range1, h_range1] += b2[v_range2, h_range2]
    return b1
```

# High pass filter

```python
def highPassFilter(img,kernel,img2_path,shapeToResize):
    stacked_image_cat = Convolve3d(img2,kernel)
    for i in range(no_of_iter):
        #print(i)
        stacked_image_cat = Convolve3d(stacked_image_cat,kernel)
    stacked_image_cat = Normalizedbetweem01(stacked_image_cat)#np.interp(stacked_image_cat, (stacked_image_cat.min(), stacked_image_cat.max()), (0, 1))
    
    #plt.imshow(stacked_image_cat)
    
    #Shape of stacked image
    shape = stacked_image_cat.shape
    # read the image
    cat = io.imread(img2_path)
    # resize to final convolve shape
    shape1 = np.asarray(cat).shape
    cat = transform.resize(cat, shapeToResize, mode='symmetric', preserve_range=True)
    cat = Normalizedbetweem01(cat)#np.interp(small_grey, (small_grey.min(), small_grey.max()), (0, 1))
#   shape1 = np.asarray(cat).shape
#    ########################################################################
#    pad = (cat.shape[0] - stacked_image_cat.shape[0])
#    pad = int(pad/2)
#    test = []
#    test = np.zeros((shape1[0],shape1[1]))
#    test1 = overlap(test,stacked_image_cat[:,:,0],pad,pad)
#     print("test1 shape",test1.shape)
#     for i in range(1,3):
#         test1 = np.dstack((test1,overlap(test,stacked_image_cat[:,:,i],pad,pad)))
#     print(test1.shape)
    #plt.imshow(cat)
    ##########################################################################
    
    cat_sharpen_image = cat - stacked_image_cat
    #cat_sharpen_image = cat - test1
    cat_sharpen_image = Normalizedbetweem01(cat_sharpen_image)
    return cat_sharpen_image
```

# Output of High pass filter

```python
cat_sharpen_image = highPassFilter(img2,kernel,img2_path,stacked_image_dog.shape)
plt.imshow(cat_sharpen_image)
plt.title("High Pass Resulting Image of cat")
plt.show()
```
