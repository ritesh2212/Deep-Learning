```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
from sklearn.datasets import make_circles
import pandas as pd
import numpy as np
import sklearn
```

## Function to plot graph based on weight ans bais

```python
def plotGraph(weight,bais,size,df):
    x = np.random.randint(-10,10, size=(size))

    y = np.multiply(-weight[0]/weight[1],x)-(bais/weight[1])

    colors = {0:'red', 1:'blue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')

    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])

    plt.plot(x,y)
    plt.show()
```

# Perceptron algorithm

```python
def perceptron(df,X,y,weight,bais,size=12,threshold=10):
    j = 0
    while(j!=threshold):
        j+=1
        for inputs,label in zip(X,y):
            #print(weight.shape," ",X.shape)
            yDash = np.dot(weight,inputs)+bais
            if yDash <= 0:
                yDash = 0
            else:
                yDash = 1
                
            weight = weight + inputs*learning_rate*(label- yDash)
            bais = bais + learning_rate*(label - yDash)
        print("Iteration : ",j," Weights : ",weight, " bais : ",bais)
        plotGraph(weight,bais,size,df)
    return weight,bais
def predict(inputs,weight,bais):
    pred = np.dot(weight,inputs)+bais
    
    #print(bais)
    if pred <= 0:
        return 0
    else:
        return 1
```

# Linear Dataset 1

```python
# # generate 2d classification dataset
X, y = make_blobs(n_samples=200, centers=2, n_features=2,shuffle=True,random_state=42)
# # scatter plot, dots colored by class value
df_Linear = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
#df.to_csv('./dataset/LinearDataset.csv',sep="\t",index=None)
#df = pd.read_csv('./dataset/LinearDataset.csv',sep='\t')
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df_Linear.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
X = df_Linear.values
y = X[:,-1]
X = X[:,:2]
# y[y==0] = -1
learning_rate = 0.01
weight = np.zeros(2)
bais = 0

weight,bais = perceptron(df_Linear,X,y,weight,bais,12)
i=10
print(weight,bais)
print(predict(X[i],weight,bais))
print(y[i])
```
