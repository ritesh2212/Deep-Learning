```python
def changeNanCategorical(dataset):
    categorical = ['Alley','FireplaceQu','Fence','MiscFeature','PoolQC','GarageQual','GarageFinish','GarageType','GarageCond','BsmtQual','BsmtCond','BsmtFinType1','BsmtFinType2','BsmtExposure']
    for col in categorical:
        dataset[col].fillna('NA', inplace=True)
    return dataset
```

```python
def changeNanNumerical(dataset,y):
    for i in range(y):
        #print(test_dataset.isna().sum().sort_values()[-18:-1].to_frame().index[i])
        col = dataset.isna().sum().sort_values()[-(y+2):-1].to_frame().index[i]
        dataset[col].fillna(dataset[col].value_counts().to_frame().index[0],inplace=True)
    return dataset
```

```python
def changeStrColInOneHotEncoding(dataset):
    for col in dataset.columns:
        if isinstance(dataset[col].unique()[0], str):
            # Get one hot encoding of columns B
            one_hot = pd.get_dummies(dataset[col],prefix=str(col+"_"))
            # Drop column B as it is now encoded
            dataset = dataset.drop(col,axis = 1)
            # Join the encoded df
            dataset = dataset.join(one_hot)
    return dataset
train_dataset = changeStrColInOneHotEncoding(train_dataset)
```

```python
train_dataset = (train_dataset - train_dataset.mean())/train_dataset.std()
train_dataset = train_dataset.values#.astype(np.float)
#train_dataset[train_dataset=='NA'] = 0

train_dataset = train_dataset.astype(np.float)
#train_dataset = (train_dataset - train_dataset.mean())/.std(train_dataset)
label = label.values.astype(np.float)

torch_tensor = torch.tensor(train_dataset)
torch_tensor = torch_tensor.type(torch.FloatTensor)

label_tensor = torch.tensor(label)
label_tensor =  label_tensor.type(torch.FloatTensor)

print(label_tensor.shape)

tensor_x,tensor_y = torch_tensor,label_tensor
my_dataset = utils.TensorDataset(tensor_x,tensor_y) # create your datset
train_loader = utils.DataLoader(my_dataset,batch_size=1) # create your dataloader
```

```python
def init_weights(m):
    if type(m) == nn.Linear:
        torch.nn.init.xavier_uniform_(m.weight)
        m.bias.data.fill_(2)
```

```python
class Net(nn.Module):
    def __init__(self,size):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(size, 1000)  
        #self.fc2 = nn.Linear(512, 128)
        #self.fc3 = nn.Linear(128, 16)
        self.fc4 = nn.Linear(1000,1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        #x = F.relu(self.fc2(x))
        #x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x


model = Net(tensor_x.shape[1])
model.apply(init_weights)
print(model)
```

```python
#model = model.cuda()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(10000):
    ave_loss=0.0
    
    for i,data in enumerate(train_loader,0):
        inputs,target =data
        #print(target)
        inputs = inputs.float()
        #print(inputs.shape)
        inputs = inputs.view(-1,302)
        #print(inputs.shape)
        target = target.float()
        #print(target.shape)
        optimizer.zero_grad()
        output=model(inputs)
        #print(output)
        output = output.squeeze(1)
        #print("Output shape",output.shape)
        #print("Target shape",target.shape)
        loss=criterion(output,target)
        loss.backward()
        optimizer.step()
        ave_loss+=loss.item()
    print("Iter ",epoch," loss : ",ave_loss/train_dataset.shape[0])
        #print(i," ",loss)
#         if i%200==0:    # print every 200 mini-batches
#             print('[%d, %5d] loss: %.3f' %
#                   (epoch + 1, i + 1, ave_loss / 100))
#             ave_loss = 0.0
```

```python
torch_tensor = torch.tensor(test_dataset)
tensor_x = torch_tensor.type(torch.FloatTensor)
my_dataset = utils.TensorDataset(tensor_x) # create your datset
test_loader = utils.DataLoader(my_dataset,batch_size=1) # create your dataloader
predict_test = []
with torch.no_grad():
    for inputs in test_loader:
        inputs=torch.from_numpy(np.array(inputs))
        outputs=model(inputs)
        predict_test.append(outputs.numpy())
```

```python
test_df = pd.DataFrame(columns=['Id', 'SalePrice'])

test_dataset1 = pd.read_csv('./dataset/kaggle/test.csv',sep=',')
test_df['Id'] = test_dataset1.Id
test_df['SalePrice'] = predict_test

test_df.to_csv("submission.csv",index=False)
```
