```python
import matplotlib.pyplot as plt
import numpy as np
```

# For 1st setting

```python
w1 = 10
w2 = 10
lamda1 = 0.02
lamda2 = 0.2
lr = 0.1
Error = []
epochs = 300
for i in range(epochs):
    E = 1/2*lamda1*w1*w1 +1/2*lamda2*w2*w2
    dEwW1 = lamda1*w1 
    dEwW2 = lamda2*w2
    w1 = w1 - lr*dEwW1
    w2 = w2 - lr*dEwW2
    Error.append(E)
epoch_range = np.arange(1,epochs+1)
plt.plot(epoch_range,Error)
plt.show()
```
